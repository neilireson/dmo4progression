```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir = os.getenv('DATA_DIR', '../Dataset/')
data_file_path = data_dir + r'MS_dataset_v.7.3.csv'
data = pd.read_csv(data_file_path, low_memory=False)
```

```python
# everyone has a T1 so we can use this to get the demographic info
data_low = data[data["EDFSCR1L"] <= 4.0]
data_medium = data[(data["EDFSCR1L"] >= 4.5) & (data["EDFSCR1L"] <= 5.0)]
data_high = data[data["EDFSCR1L"] >= 5.5]
```

```python
def demographics_table(data_all):
    data = data_all[data_all["visit.number"] == "t1"].reset_index(drop=True)
    PDDS_lookup = {
        "Normal":0,
        "Mild Disability":1,
        "Moderate Disability":2,
        "Gait Disability":3,
        "Early Cane":4,
        "Late Cane":5,
        "Bilateral Support":6,
        "Wheelchair/Scooter":7,
        "Bedridden":8
    }

    demographics = {}
    demographics["count"] = len(data)
    demographics["Age, mean (SD)"] = f"{data["age"].mean():.1f} ({data["age"].std():.1f})"
    demographics["Gender, female, n (%)"] = f"{len(data[data["gender"] == "Female"])} ({100*(len(data[data["gender"] == "Female"])/len(data)):.1f})"
    college_higher = len(data[data["edhigh"] == "Undergraduate/Bachelor"]) + len(data[data["edhigh"] == "Postgraduate (Diploma, Master etc)"]) + len(data[data["edhigh"] == "Doctoral/PhD or higher"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] > 15)])
    high_school = len(data[data["edhigh"] == "High school leaving certificate/diploma or equivalent (11 or more years)"]) + len(data[data["edhigh"] == "Tertiary education (e.g. trade or apprenticeship)"]) + len(data[data["edhigh"] == "College/University but no degree"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] > 10) & (data["edyears"] <= 15)])
    less_than_high_school = len(data[data["edhigh"] == "Less than high school (0-10 years of schooling)"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] <= 10)])
    demographics["Education, n (%)"] = {
        "College degree or higher": f"{college_higher} ({100*(college_higher/len(data)):.1f})",
        "High school": f"{high_school} ({100*(high_school/len(data)):.1f})",
        "Less than high school": f"{less_than_high_school} ({100*(less_than_high_school/len(data)):.1f})"
    }
    demographics["EDSS, median (IQR)"] = f"{data["EDFSCR1L"].median()} ({data["EDFSCR1L"].quantile(0.75) - data["EDFSCR1L"].quantile(0.25)})"
    demographics["MSWS-12, mean (SD)"] = f"{data["MSWTOT1N"].mean():.1f} ({data["MSWTOT1N"].std():.1f})"
    demographics["T25-FW, mean (SD)"] = f"{data["TFWMN1N"].mean():.1f} ({data["TFWMN1N"].std():.1f})"
    demographics["PDDS, median (IQR)"] = f"{data["PDDSW1L"].map(PDDS_lookup).median():.1f} ({(data["PDDSW1L"].map(PDDS_lookup).quantile(0.75) - data["PDDSW1L"].map(PDDS_lookup).quantile(0.25)):.1f})"
    demographics["MS Subtype"] = {
        "Primary Progressive, n": len(data[data["mstype"] == "Primary Progressive MS"]),
        "Secondary Progressive, n": len(data[data["mstype"] == "Secondary Progressive MS"]),
        "Relapse Remitting, n": len(data[data["mstype"] == "Relapsing Remitting MS"])
    }
    demographics["Indoor Walking Aid, n (%)"] = f"{len(data[data["aidindoor"] == True])} ({100*(len(data[data["aidindoor"] == True])/len(data)):.1f})"
    demographics["Outdoor Walking Aid, n (%)"] = f"{len(data[data["aidoutdoor"] == True])} ({100*(len(data[data["aidoutdoor"] == True])/len(data)):.1f})"
    demographics["Falls in the previous 12 months, n"] = f"{data_all["fallsnum"].dropna().astype(int).sum()}"
    demographics["6 min walk distance, mean (SD)"] = f"{data_all["6mwd1"].mean():.1f} ({data_all["6mwd1"].std():.1f})"
    demographics["Systolic BP, mmHg, mean (SD)"] = f"{data["bpsyst"].mean():.1f} ({data["bpsyst"].std():.1f})"

    return demographics

display(demographics_table(data))
display(demographics_table(data_low))
display(demographics_table(data_medium))
display(demographics_table(data_high))
```
