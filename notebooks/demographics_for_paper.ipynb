```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir = os.getenv('DATA_DIR', '../Dataset/')
data_file_path = data_dir + r'MS_dataset_v.7.3.csv'
data = pd.read_csv(data_file_path, low_memory=False)
```

```python
# everyone has a T1 so we can use this to get the demographic info
data = data[data["visit.number"] == "t1"].reset_index(drop=True)
data_low = data[data["EDFSCR1L"] <= 4.0]
data_medium = data[(data["EDFSCR1L"] >= 4.5) & (data["EDFSCR1L"] <= 5.0)]
data_high = data[data["EDFSCR1L"] >= 5.5]
```

```python
def demographics_table(data):
    demographics = {}
    demographics["count"] = len(data)
    demographics["age"] = data["age"].mean()
    # this is a mess
    demographics["education"] = {
        "College degree or higher": len(data[data["edhigh"] == "Undergraduate/Bachelor"]) + len(data[data["edhigh"] == "Postgraduate (Diploma, Master etc)"]) + len(data[data["edhigh"] == "Doctoral/PhD or higher"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] > 15)]),
        "High school": len(data[data["edhigh"] == "High school leaving certificate/diploma or equivalent (11 or more years)"]) + len(data[data["edhigh"] == "Tertiary education (e.g. trade or apprenticeship)"]) + len(data[data["edhigh"] == "College/University but no degree"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] > 10) & (data["edyears"] <= 15)]),
        "Less than high school": len(data[data["edhigh"] == "Less than high school (0-10 years of schooling)"]) + len(data[(data["edhigh"] == "Other") & (data["edyears"] <= 10)])
    }
    demographics["EDSS median (IQR)"] = f"{data["EDFSCR1L"].median()} ({data["EDFSCR1L"].quantile(0.75) - data["EDFSCR1L"].quantile(0.25)})"
    demographics["MSWS-12 mean (SD)"] = f"{data["MSWTOT1N"].mean():.2f} ({data["MSWTOT1N"].std():.2f})"
    return demographics

display(data_low["edhigh"].unique())

display(demographics_table(data_low))
display(demographics_table(data_medium))
display(demographics_table(data_high))
```
